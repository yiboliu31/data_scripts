```python
%matplotlib inline
import matplotlib.pyplot as plt
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval
import numpy as np
import skimage.io as io
import pylab
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
annType = ['segm','bbox','keypoints']
annType = annType[1]      #specify type here
prefix = 'person_keypoints' if annType=='keypoints' else 'fixed_instances'
print('Running demo for *%s* results.'%(annType))
```

```python
#initialize COCO ground truth api
dataDir='/media/dean/My Passport for Mac/Kache/datasets/darknet/data/coco'
dataType='val2014'
annFile = '%s/annotations/%s_%s.json'%(dataDir,prefix,dataType)
cocoGt=COCO(annFile)
```

```python
#initialize COCO detections api
resFile='/home/dean/Desktop/coco_results.json'
cocoDt=cocoGt.loadRes(resFile)
```

```python
categories = cocoDt.loadCats(cocoDt.getCatIds())
category_names = [category['name'] for category in categories]
print(sorted(category_names))
category_names = list(set(category_names) - set(['car','truck']))
print('/n/n/n',sorted(category_names))

category_ids = cocoDt.getCatIds(catNms=category_names)
```

## Pretrained Results

```python
#initialize COCO detections api
resFile='/home/dean/Desktop/pretrain_coco_results.json'
cocoDt=cocoGt.loadRes(resFile)

for cat in category_names:
    category_ids = cocoDt.getCatIds(catNms=cat)
    imgIds=cocoDt.getImgIds(catIds=category_ids)
    print('MAP for:', cat)
    # running evaluation
    cocoEval = COCOeval(cocoGt,cocoDt,annType)
    cocoEval.params.imgIds  = imgIds
    cocoEval.evaluate()
    cocoEval.accumulate()
    cocoEval.summarize()
    print('\n\n\n')
```
